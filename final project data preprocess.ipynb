```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Base.csv')
```

```python
df.head()
```

```python
pd.set_option('display.max_columns', None)

df.describe()
```

```python
columns_to_standardize = ['income','name_email_similarity','current_address_months_count','customer_age','days_since_request','intended_balcon_amount','zip_count_4w','velocity_6h', 'velocity_24h' ,'velocity_4w','bank_branch_count_8w', 'date_of_birth_distinct_emails_4w','credit_risk_score','bank_months_count','proposed_credit_limit','session_length_in_minutes']
```

```python
df.drop('prev_address_months_count', axis=1, inplace=True)
df.drop('device_fraud_count', axis=1, inplace=True)
```

```python
scaler = StandardScaler()


# Fit the scaler to the columns and transform them
df[columns_to_standardize] = scaler.fit_transform(df[columns_to_standardize])
df
```

```python
one_hot_encoded_df = pd.get_dummies(df, columns=['payment_type','employment_status','housing_status','source','device_os','device_distinct_emails_8w','month'])
one_hot_encoded_df
```

```python
one_hot_encoded_df.describe()
```

```python
one_hot_encoded_df.to_csv('base_cleaned.csv', index=False)
```
